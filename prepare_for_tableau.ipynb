```python
import pandas as pd
```

```python
df = pd.read_csv('recleaned_df.csv')
```

```python
df.columns
```

```python
df['province'].unique()
```

```python
df['municipality'].nunique()
```

```python
df = df[df['province'] == 'Madrid']
```

```python
df
```

```python
bin_edges = list(range(0, int(df['price'].max()) + 200000, 200000))
bin_labels = [f"{i}-{i+200000}" for i in bin_edges[:-1]]
df['price_range'] = pd.cut(df['price'], bins=bin_edges, labels=bin_labels, include_lowest=True)
```

```python
df[['price_range', 'price']].head()
```

```python
bin_edges = list(range(0, int(df['size'].max()) + 150, 150))
bin_labels = [f"{i}-{i+150}" for i in bin_edges[:-1]]
df['size_range'] = pd.cut(df['size'], bins=bin_edges, labels=bin_labels, include_lowest=True)
```

```python
df[['size_range', 'size']]
```

```python
df['municipality'].unique()
```

```python
df.to_csv('prepared.csv')
```
